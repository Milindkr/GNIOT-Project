# Q. Program that inputs three numbers and calculates two sums as per this
###	Sum1 as the sum of all inputs numbers
###	Sum 2 as the sum of non-duplicates numbers; if there are duplicate numbers in the input, 	ignores them

```python
num1 = int(input("Enter 1st num "))
num2 = int(input("Enter 2nd num"))
num3 = int(input("Enter 3rd num"))

sum1 = num1+num2+num3
sum2 = 0

if num1 != num2 and num1 != num3:
  sum2 = sum2 + num1
if num2 != num1 and num2 != num3:
  sum2 += num2
if num3 != num1 and num3 != num2:
  sum2 +=num3

print(sum1)
if sum2==0:
  print("All the numbers are same")
else:
  print(sum2)
```
