##Arbitrary Arguments - *args

add  * before the parameter name

```python
def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")
```

##Arbitrary Keyword Arguments - **kwargs
add two asterisks ** before the parameter name.

```python
def my_function(**kid):
  print("His name is " + kid["fname"])

my_function(fname= "Tobias")
```

##Combining *args and **kwargs
- You can use both *args and **kwargs in the same function.

- The order must be:

- regular parameters
- *args
- **kwargs

```python
def my_function(title, *args, **kwargs):
  print("Title:", title)
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

my_function("User Info", "Emil", "Tobias", age = 25, city = "Oslo")
```

##what is variable length argument

- A variable-length argument in Python allows a function to accept any number of arguments, instead of a fixed number.
- Sometimes you don’t know how many inputs a function will receive.
Variable-length arguments make your function flexible and reusable.

```python
def add_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(1, 2, 3))       # 6
print(add_numbers(10, 20, 30, 40)) # 100
```

##Recursion

- Recursion is when a function calls itself.

---------
- A base case - A condition that stops the recursion
- A recursive case - The function calling itself with a modified argument



```python
#Print Numbers from N to 1

def print_desc(n):
  #base case
    if n == 0:
        return
  #rec case
    print(n)
    print_desc(n - 1)

print_desc(5)
```

```python
#Sum of First N Natural Numbers

def sum(n):
  #base case
  if n==0:
    return 0

  #recursive case
  return n+sum(n-1)

sum(5)
```

```python
def factorial(n):
  #base case
  if n==0 or n==1:
    return 1
  #recursive case
  return n*factorial(n-1)


factorial(3)
```

```python
def is_palindrome(s):

  #Base case
    if len(s) <= 1:
        return True

    if s[0] != s[-1]:
        return False

    #recursive case
    return is_palindrome(s[1:-1])

print(is_palindrome("h"))
```

##Decorators

A decorator is a function that takes another function as input and returns a new function.

```python
def changecase(fun):
  def innercase():
    return fun().upper()
  return innercase

@changecase
def name():
  return "Hello Sally"

print(name())
```
