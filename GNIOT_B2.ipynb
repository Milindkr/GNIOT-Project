# Determine ticket price
#age<5 = free
#age<=12 = student 50 not student = 70
#age<= 17 student 125 not student =150
#age <=64 student = 130 not student 170
#else = 180

```python
age  = int(input("Enter your age"))
is_student = input("is student ? (yes/No): ").lower()

if age <= 5:
  price = "Free"
elif age<= 12:
  if is_student == 'yes':
    price = "50"
  else:
    price = "70"
elif age <=17:
  if is_student == 'yes':
    price = "125"
  else:
    price = "150"
elif age <= 64:
  if is_student == 'yes':
    price = "130"
  else:
    price = "170"
else:
  price = "180"

print(price)
```

```python
#What are Keyword Argument

def _function(animal,owner):
  print("Hi",owner, "owner of", animal)

_function(owner="Mayank",animal="CAT")
```

```python
#,/

def _function(animal,owner,/):
  print("Hi",owner, "owner of", animal)

_function("Mayank","CAT")
```

```python
#*,

def _function(*,animal,owner):
  print("Hi",owner, "owner of", animal)

_function(owner="Mayank",animal="CAT")
```

```python
def add_num(*args):
  total_sum = 0
  for num in args:
    total_sum += num
  return total_sum

print(add_num(10,20,30,40))
```
