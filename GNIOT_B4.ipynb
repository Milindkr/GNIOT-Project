```python
## Simple Calculator program


a = int(input("Enter num1 "))
b = int(input("Enter num2 "))
operation = input("Enter operation (+,-,*,/,%,//)")

if operation == '+':
  print(a+b)
elif operation == "-":
  print(a-b)
elif operation == "*":
  print(a*b)
elif operation == "/":
  if b != 0:
    print(a/ b)
  else:
    print("Errpr! Division by zero")
elif operation == "%":
  print(a%b)
else:
  print("Out of Operation")
```
